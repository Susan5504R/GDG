```python
## 1. Data Preparation and Splitting (Corrected with Label Encoding)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# --- Load the Dataset and Parse Extensions (if in a new notebook) ---
DATA_PATH = '../data/final_dataset.csv'
df = pd.read_csv(DATA_PATH)

def parse_extensions(ext_string):
    try:
        cleaned_string = ext_string.replace("np.str_('", "'").replace("')", "'")
        ext_list = ast.literal_eval(cleaned_string)
        return ext_list if isinstance(ext_list, list) else [ext_list]
    except (ValueError, SyntaxError):
        return []
df['extensions_list'] = df['fileextensions'].apply(parse_extensions)

# --- Define Features (X) and Target (y) ---
features = [
    'committype', 'numfileschanged', 'linesadded', 
    'linesdeleted', 'numcommentsadded', 'commitmessage', 'extensions_list'
]
X = df[features]
y = df['role']

# --- Encode the Target Variable (y) ---
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Print the mapping to see what the numbers mean
print("Label Encoding Mapping:")
for i, class_name in enumerate(le.classes_):
    print(f"{class_name} -> {i}")

# --- Split the data using the encoded target ---
X_train, X_test, y_train_encoded, y_test_encoded = train_test_split(
    X, 
    y_encoded, # Use the encoded y
    test_size=0.2,
    random_state=42,
    stratify=y_encoded # Stratify using the encoded y
)

print(f"\nTraining set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")
```

```python
## 2. Build a Comprehensive Preprocessing Pipeline (Corrected)

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, FunctionTransformer

# --- Define the preprocessing steps for each data type ---

numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
text_transformer = TfidfVectorizer(stop_words='english', max_features=1000)

# This is a mini-pipeline: first join the list into a string, then apply TF-IDF.
extensions_transformer = Pipeline(steps=[
    # The fix is to add `feature_names_out='one-to-one'` here
    ('joiner', FunctionTransformer(lambda x: x.apply(' '.join), feature_names_out='one-to-one')),
    ('tfidf', TfidfVectorizer(max_features=50))
])

# --- Define which columns each transformer should apply to ---
numeric_features = ['numfileschanged', 'linesadded', 'linesdeleted', 'numcommentsadded']
categorical_features = ['committype']
text_features = 'commitmessage'
extension_features = 'extensions_list'

# --- Assemble the preprocessor with ColumnTransformer ---
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features),
        ('text', text_transformer, text_features),
        ('ext', extensions_transformer, extension_features)
    ],
    remainder='drop'
)

print("Preprocessor definition has been corrected and updated.")
```

```python
## 3. Train and Evaluate Baseline Model (Logistic Regression)

import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, f1_score, ConfusionMatrixDisplay

# --- Create the full pipeline by attaching the classifier to the preprocessor ---
baseline_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=42, max_iter=1000)) # Increased max_iter for convergence
])

# --- Train the entire pipeline on the training data ---
print("Training the baseline model...")
baseline_model.fit(X_train, y_train)
print("Training complete.")

# --- Make predictions on the unseen test data ---
y_pred_base = baseline_model.predict(X_test)

# --- Evaluate the model's performance ---
print("\n--- Baseline Model: Logistic Regression ---")

# Print the detailed classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_base))

# Calculate and print the primary metric: Macro F1 Score
macro_f1_base = f1_score(y_test, y_pred_base, average='macro')
print(f"Primary Metric (Macro F1 Score): {macro_f1_base:.4f}")

# Plot the confusion matrix to visualize errors
print("\nConfusion Matrix:")
ConfusionMatrixDisplay.from_estimator(baseline_model, X_test, y_test, xticks_rotation='vertical')
plt.title('Baseline Model Confusion Matrix')
plt.show()
```

```python
## 4. Train and Evaluate Improved Model (LightGBM)

import lightgbm as lgb

# --- Create the improved model pipeline ---
# The only change is the classifier at the end.
improved_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', lgb.LGBMClassifier(random_state=42))
])

# --- Train the entire pipeline on the training data ---
print("Training the improved model (LightGBM)...")
improved_model.fit(X_train, y_train)
print("Training complete.")

# --- Make predictions on the unseen test data ---
y_pred_improved = improved_model.predict(X_test)

# --- Evaluate the model's performance ---
print("\n--- Improved Model: LightGBM ---")

# Print the detailed classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_improved))

# Calculate and print the primary metric: Macro F1 Score
macro_f1_improved = f1_score(y_test, y_pred_improved, average='macro')
print(f"Primary Metric (Macro F1 Score): {macro_f1_improved:.4f}")

# Plot the confusion matrix to visualize errors
print("\nConfusion Matrix:")
ConfusionMatrixDisplay.from_estimator(improved_model, X_test, y_test, xticks_rotation='vertical')
plt.title('Improved Model Confusion Matrix')
plt.show()
```

```python
## Optional Experiment: XGBoost (Corrected)

import xgboost as xgb

# --- Create the XGBoost model pipeline ---
xgb_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', xgb.XGBClassifier(
        random_state=42, 
        use_label_encoder=False, 
        eval_metric='mlogloss'
    ))
])

# --- Train the model using the ENCODED training labels ---
print("Training the XGBoost model...")
xgb_model.fit(X_train, y_train_encoded) # Use y_train_encoded
print("Training complete.")

# --- Evaluate the model ---
y_pred_xgb = xgb_model.predict(X_test)

# In the report, we use the encoded test labels and provide the original names
print(classification_report(y_test_encoded, y_pred_xgb, target_names=le.classes_))

macro_f1_xgb = f1_score(y_test_encoded, y_pred_xgb, average='macro')
print(f"\nXGBoost Model (Macro F1 Score): {macro_f1_xgb:.4f}")

# For the confusion matrix, we provide the display labels
ConfusionMatrixDisplay.from_estimator(
    xgb_model, 
    X_test, 
    y_test_encoded, # Use encoded y
    xticks_rotation='vertical', 
    display_labels=le.classes_ # Use original class names for the plot
)
plt.title('XGBoost Model Confusion Matrix')
plt.show()
```

```python
## 5. Model Comparison (Updated)

# Create a dictionary to hold the final results from all experiments
results = {
    'Model': [
        'Baseline (Logistic Regression)', 
        'Improved (LightGBM)', 
        'Optional (XGBoost)'
    ],
    'Macro F1 Score': [
        macro_f1_base, 
        macro_f1_improved, 
        macro_f1_xgb
    ]
}

# Create a pandas DataFrame from the dictionary for a clean presentation
results_df = pd.DataFrame(results).sort_values(by='Macro F1 Score', ascending=False)

# Display the final results table
print("--- Model Performance Comparison ---")
display(results_df)
```

```python
## 6. Model Interpretation and Feature Importance
import seaborn as sns
# --- Extract feature names and importances ---
# Get the trained classifier step from the improved model pipeline
lgbm_classifier = improved_model.named_steps['classifier']

# Get the feature names from the preprocessor step
# This correctly gets names for one-hot encoded, TF-IDF, and numeric features
feature_names = improved_model.named_steps['preprocessor'].get_feature_names_out()
feature_importances = lgbm_classifier.feature_importances_

# --- Create a DataFrame for visualization ---
importance_df = pd.DataFrame({
    'feature': feature_names,
    'importance': feature_importances
}).sort_values('importance', ascending=False)


# --- Plot the top 25 most important features ---
plt.figure(figsize=(12, 10))
sns.barplot(
    x='importance',
    y='feature',
    data=importance_df.head(25),
    palette='rocket'
)
plt.title('Top 25 Most Important Features for Role Prediction', fontsize=16)
plt.xlabel('Importance Score', fontsize=12)
plt.ylabel('Feature Name', fontsize=12)
plt.show()
```

```python
## 8. Hyperparameter Tuning with RandomizedSearchCV

from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform

# --- Define the hyperparameter search space ---
# We tell the search what settings to try for the classifier step of our pipeline.
# Note the 'classifier__' prefix, which targets a step in the pipeline.
param_dist = {
    'classifier__n_estimators': randint(150, 500),
    'classifier__learning_rate': uniform(0.01, 0.2),
    'classifier__num_leaves': randint(20, 50),
    'classifier__max_depth': randint(5, 15),
}

# --- Set up the Randomized Search with Cross-Validation ---
random_search = RandomizedSearchCV(
    estimator=improved_model,          # Our LightGBM pipeline
    param_distributions=param_dist,
    n_iter=25,                         # How many random combinations to try
    cv=3,                              # 3-fold cross-validation
    scoring='f1_macro',                # The metric to optimize for!
    random_state=42,
    n_jobs=-1,                         # Use all available CPU cores to speed it up
    verbose=1                          # Show progress
)

# --- Run the search ---
print("Starting hyperparameter tuning...")
random_search.fit(X_train, y_train)
print("Tuning complete.")

# --- Get the best model and evaluate it ---
print("\n--- Best Hyperparameters Found ---")
print(random_search.best_params_)

best_tuned_model = random_search.best_estimator_
y_pred_tuned = best_tuned_model.predict(X_test)

print("\n--- Tuned Model Performance ---")
print(classification_report(y_test, y_pred_tuned))

macro_f1_tuned = f1_score(y_test, y_pred_tuned, average='macro')
print(f"Final Tuned Model (Macro F1 Score): {macro_f1_tuned:.4f}")

# You can now update your comparison table with this new, best score.
results_df.loc[2] = ['Tuned (LightGBM)', macro_f1_tuned]
display(results_df)
```
